import numpy as np

```python
a = np.array([[1, 6],
             [2, 8],
             [3, 11],
             [3, 10],
             [1, 7]])
a
```

```python
mean_a = np.array([np.mean(a[:,0]), np.mean(a[:,1])])
mean_a
```

# Задание 2

```python
a_centered = a - mean_a
a_centered
```

# Задание 3

```python
a_centered_sp = np.dot(a_centered[:,0], a_centered[:, 1])
N = a.shape[0]
a_centered_sp/(N-1)
```
